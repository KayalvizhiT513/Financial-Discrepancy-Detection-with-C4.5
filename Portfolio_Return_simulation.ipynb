```python
import pandas as pd
import numpy as np
from collections import Counter
import math

def log(p):
    if p != 0:
        return math.log2(p)
    else:
        return 0

# Function to calculate entropy
def entropy_before_split(p):

    return -(p * log(p)) - ((1-p) * log(1-p))

def entropy_after_split(n, x, y, px, fx, py, fy):
    if x == 0:
        left_entropy = 0
    else:
        left_entropy = -(x/n)*(((px/x) * log(px/x)) + ((fx/x) * log(fx/x)))
    if y == 0:
        right_entropy = 0
    else:
        right_entropy = -(y/n)*(((py/y) * log(py/y)) + ((fy/y) * log(fy/y)))
    return left_entropy + right_entropy

# Function to calculate information gain
def information_gain(entopy_before_split, entropy_after_split):
    return entopy_before_split - entropy_after_split

# Function to calculate split information
def split_info(x, y, n):
    return -((x/n) * log(x/n)) - ((y/n) * log(y/n))

# Function to calculate gain ratio
def gain_ratio(info_gain, si):
    return info_gain / si if si != 0 else 0
```

```python
columns = ['Pass:Fail', 'Before split True', 'Before split False', 'True After split Pass', 'True After split Fail', 'False After split Pass', 'False After split Fail', 'Entropy before split', 'Entropy after split', 'Information gain', 'Split info', 'Gain ratio']
simulation_df = pd.DataFrame(columns=columns)
n = 100

for m in range(0, n+1):
    for i in range(0, n+1): #split
        x = i
        y = n-i
        for j in range(0, i+1):  # Pass Fail before split
            px = j
            fx = x - px if x > px else 0
            py = m - px if m > px and n > x else 0
            fy = y - py if y > py and n > x+py else 0
            before_split = entropy_before_split(x/n)
            after_split = entropy_after_split(n, x, y, px, fx, py, fy)
            info_gain = information_gain(before_split, after_split)
            si = split_info(x, y, n)
            if x == 93 and y == 7:
                simulation_df = simulation_df.append(pd.Series([f'{m}:{abs(n-m)}',x, y, px, fx, py, fy, before_split, after_split, info_gain, si, gain_ratio(info_gain, si)], index=simulation_df.columns), ignore_index=True)
```
